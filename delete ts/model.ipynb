```python
import pandas as pd
```

```python
df = pd.read_csv('preprocessed_telco_churn.csv')
df = df[['SeniorCitizen','tenure','InternetService','StreamingTV','StreamingMovies','Contract','PaymentMethod','MonthlyCharges','TotalCharges','PaperlessBilling','Churn']]
df.head()
```

```python
df2 = pd.read_csv('preprocessed_raw.csv')
df2 = df2[['SeniorCitizen','tenure','InternetService','StreamingTV','StreamingMovies','Contract','PaymentMethod','MonthlyCharges','TotalCharges','PaperlessBilling','Churn']]
df2.head()
```

```python
df_final = pd.concat([df,df2],ignore_index=True)
df_final.tail()
```

```python
features = df_final[['SeniorCitizen','tenure','InternetService','StreamingTV','StreamingMovies','Contract','PaymentMethod','MonthlyCharges','TotalCharges']]
target = df_final['Churn']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


model = LogisticRegression(max_iter=756)
x_train,x_test,y_train,y_test = train_test_split(features,target,test_size=0.3,random_state=1)

model.fit(x_train,y_train)

y_pred = model.predict(x_test)

model.score(x_test,y_test)
```
